```python
#Example 3 adapted from #https://stackoverflow.com/questions/47094949/labeling-edges-in-networkx

edges = [['Conrad','Mary'], ['Conrad','Cindy'], ['Conrad','Alice'],['Alice','Brad'],
         ['Alice','Angie'], ['Alice','Jim'],['Cindy','Samuel'], ['Cindy','Dave'], ['Cindy','Frida']]
G = nx.Graph()
G.add_edges_from(edges)
pos = nx.spring_layout(G)
plt.figure()
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    labels={node: node for node in G.nodes()}
)
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={('Conrad','Mary'): 'CM', 
                 ('Conrad','Cindy'): 'CC', 
                 ('Conrad','Alice'): 'CA',
                 ('Alice','Brad'): 'AB',
                 ('Alice','Angie'): 'AA',
                 ('Alice','Jim'): 'AJ',
                 ('Cindy','Samuel'): 'CS',
                 ('Cindy','Dave'): 'CD',
                 ('Cindy','Frida'): 'CF'},
    font_color='red'
)
plt.axis('off')
plt.show()
```
